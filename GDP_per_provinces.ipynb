```python
import matplotlib.pyplot as plt

# Group data by province and year
df_province_year = df.groupby(['GEO', 'YEAR'])['VALUE'].sum().unstack()

# Create subplots for each province
num_provinces = len(df_province_year.index)
num_rows = (num_provinces + 2) // 3  # Adjust number of rows as needed
num_cols = min(3, num_provinces)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
fig.tight_layout(pad=5.0)

for i, province in enumerate(df_province_year.index):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col] if num_rows > 1 else axes[col]
    ax.plot(df_province_year.columns, df_province_year.loc[province])
    ax.set_title(province)
    ax.set_xlabel('Year')
    ax.set_ylabel('GDP Value')
    # Set background color to black for each subplot
    ax.set_facecolor('black')

# Remove any unused subplots
for i in range(num_provinces, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    if num_rows > 1:
        fig.delaxes(axes[row, col])
    else:
        fig.delaxes(axes[col])

plt.show()
```

```python
# Filter the sum of value for every industry
df_industry = df.groupby('North American Industry Classification System (NAICS)')['VALUE'].sum()
df_industry = df_industry.sort_values(ascending=False)
df_industry
```

```python
# Applying pareto rule with df_industry

import pandas as pd

# Assuming df_industry is your DataFrame with industry and GDP value
def pareto_rule_industry(df_industry):
  """Applies the Pareto rule (80/20 rule) to identify the most important industries.

  Args:
    df_industry: A pandas DataFrame with 'VALUE' as the GDP value and the index
      as the industry classification.

  Returns:
    A tuple containing:
      - A DataFrame with the top industries contributing to 80% of the total GDP.
      - The percentage of total GDP represented by the top industries.
  """

  df_industry = df_industry.sort_values(ascending=False)
  total_gdp = df_industry.sum()
  cumulative_gdp = df_industry.cumsum()
  pareto_threshold = 0.8 * total_gdp
  top_industries = df_industry[cumulative_gdp <= pareto_threshold]
  percentage_gdp = (top_industries.sum() / total_gdp) * 100

  return top_industries, percentage_gdp


# Apply the Pareto rule to df_industry
top_industries, percentage_gdp = pareto_rule_industry(df_industry)

print("Top Industries Contributing to 80% of GDP:")
print(top_industries)
print("\nPercentage of total GDP represented by the top industries:", percentage_gdp)
```

```python
# Remove the first three rows from top_industries

# Assuming top_industries is the DataFrame

top_industries = top_industries.iloc[3:]
print(top_industries)
```

```python
# Generate a graph based on the established Pareto rule.

import plotly.express as px

# Assuming top_industries is the DataFrame obtained from pareto_rule_industry
fig = px.bar(top_industries, x=top_industries.index, y='VALUE',
             labels={'VALUE': 'GDP Value', 'North American Industry Classification System (NAICS)': 'Industries'},
             title='Top Industries Contributing to 80% of GDP',
             template='plotly_dark')

fig.update_layout(plot_bgcolor='black', paper_bgcolor='black')
fig.show()
```

```python
# Filtrer le DataFrame pour exclure les lignes où NAICS est "All Industries"
df_filtered = df[df['North American Industry Classification System (NAICS)'] != 'All Industries']

# Verify the result
df_filtered
```

```python
# Filter the DataFrame to exclude rows where NAICS contains "All Industries"
df_filtered = df[~df['North American Industry Classification System (NAICS)'].str.contains('All Industries', case=False, na=False)]

# Verify the result
df_filtered
```

```python
import plotly.express as px

# Calculate the sum of GDP by province and industry
df_province_industry = df_filtered.groupby(['GEO', 'North American Industry Classification System (NAICS)'])['VALUE'].sum().reset_index()

# Find the industry with the most important GCP per provinces
top_industries = df_province_industry.loc[df_province_industry.groupby('GEO')['VALUE'].idxmax()]

# Create a treemap
fig = px.treemap(
    top_industries,
    path=['GEO', 'North American Industry Classification System (NAICS)','VALUE'],
    values='VALUE',
    title='Industry with the highest GDP by Province'
)

# Formatting
fig.update_traces(textinfo='label+value')  # Show the label and the Value
fig.update_layout(plot_bgcolor='black', paper_bgcolor='black')
fig.show()
```

```python
import plotly.express as px

# Calculate amounts by province and industry
df_industry_growth = df_filtered.groupby(['GEO', 'North American Industry Classification System (NAICS)', 'YEAR'])['VALUE'].sum().reset_index()

# Calculate the change in GDP value from one year to the next
df_industry_growth['VALUE_PREV'] = df_industry_growth.groupby(['GEO', 'North American Industry Classification System (NAICS)'])['VALUE'].shift(1)
df_industry_growth['GROWTH'] = df_industry_growth['VALUE'] - df_industry_growth['VALUE_PREV']

# Filter to keep only industries with positive growth
df_growth_positive = df_industry_growth[df_industry_growth['GROWTH'] > 0]

# Calculate total growth by province and industry
total_growth = df_growth_positive.groupby(['GEO', 'North American Industry Classification System (NAICS)'])['GROWTH'].sum().reset_index()

# Calculate total growth by province
total_growth_value = total_growth.groupby('GEO')['GROWTH'].sum().reset_index().rename(columns={'GROWTH': 'TOTAL_GROWTH'})

# Merge totals with original DataFrame
total_growth = total_growth.merge(total_growth_value, on='GEO')

# Calculate the cumulative share
total_growth['CUMULATIVE_VALUE'] = total_growth.groupby('GEO')['GROWTH'].cumsum()
total_growth['CUMULATIVE_PERCENTAGE'] = total_growth['CUMULATIVE_VALUE'] / total_growth['TOTAL_GROWTH']

# Filter to keep industries that represent 80% of total growth by province
df_pareto = total_growth[total_growth['CUMULATIVE_PERCENTAGE'] <= 0.8]

# Create the treemap with the filtered data
fig = px.treemap(
    df_pareto,
    path=['GEO', 'North American Industry Classification System (NAICS)'],
    values='GROWTH',
    title='Industries ayant connu une croissance représentant 80 % de la croissance totale par Province'
)

# Formatting
fig.update_layout(plot_bgcolor='black', paper_bgcolor='black')
fig.show()
```
