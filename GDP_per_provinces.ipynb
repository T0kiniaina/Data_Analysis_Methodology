```python
# Install the API stats_can to get a data from statcanada
!pip install stats_can

# limit the number of decimal places to two
from stats_can import StatsCan
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format


# Get GDP data for Canada's provinces
sc = StatsCan()
df = sc.table_to_df("36-10-0402-01")
df
```

```python
# Transpose the table to see all the columns

m = df.shape[1]
df0 = df.iloc[0:5, 0:m].T
df0
```

## The most important provinces


---



```python
df["YEAR"] = df["REF_DATE"].dt.year
df
```

```python
df_imp_provinces = df.groupby("GEO")["VALUE"].sum()
df_imp_provinces = df_imp_provinces.sort_values(ascending=False)
df_imp_provinces
```

```python
# prompt: build a bar plot form plotly express arrière plan noir

import plotly.express as px

# Assuming df_imp_provinces is your DataFrame with province and GDP value
fig = px.bar(df_imp_provinces, x=df_imp_provinces.index, y='VALUE',
             labels={'VALUE': 'GDP Value', 'GEO': 'Provinces'},
             title='GDP Value of Major Canadian Provinces',
             template='plotly_dark')

fig.update_layout(plot_bgcolor='black', paper_bgcolor='black')
fig.show()
```
